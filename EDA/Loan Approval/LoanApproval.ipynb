```python
# Converting the ApplicationDate column to DATEtype
df['ApplicationDate'] = pd.to_datetime(df['ApplicationDate'])
```

```python
# Columns of Interest

#Defining the columns you want to keep
columns_of_Interest = ['ApplicationDate', 'Age', 'AnnualIncome', 'CreditScore',
       'EmploymentStatus', 'LoanAmount','HomeOwnershipStatus', 'SavingsAccountBalance','TotalAssets',
         'TotalLiabilities','JobTenure', 'LoanApproved']

# Create a new DataFrame with only the specified columns
loans = df[columns_of_Interest]

loans.head()
```

```python
# Check the shape
loans.shape
```

## Data Exploration and Visualization

```python
# Checking on the data points of interest
loans.columns
```

    Applications over Time

```python
loans.tail(2)
```

```python
loans = loans.iloc[:-1]
loans.tail()
```

```python
# Extract the year and month from the 'ApplicationDate'
loans['Year'] = loans['ApplicationDate'].apply(lambda x : str(x).split('-')[0])
loans['Month'] = loans['ApplicationDate'].apply(lambda x : str(x).split('-')[1])
```

```python
# loans.loc[:, 'Year'] = loans['ApplicationDate'].apply(lambda x : str(x).split('-')[0])
# loans.loc[:, 'Month'] = loans['ApplicationDate'].apply(lambda x : str(x).split('-')[1])
```

```python
loans.head()
```

```python
# Group by year and count applications
applications_per_year = loans.groupby(['Year','LoanApproved'], as_index=False)['LoanApproved'].value_counts()

plt.figure(figsize=(10,12))

sns.barplot(data = applications_per_year, y= 'Year', x= 'count', hue= 'LoanApproved')
plt.show()
```
