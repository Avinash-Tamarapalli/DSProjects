    Annual Income

```python
##AnnualIncome
#ApplicationDate
#EmploymentStatus
#LoanAmount
#HomeOwnershipStatus
#LoanApproved

# Measures of Central Tendency

mean_AI = round(loans['AnnualIncome'].mean(),2)
mode_AI = round(loans['AnnualIncome'].mode())
median_AI = round(loans['AnnualIncome'].median(),2)

'''.quantile(0.25): This calculates the 25th percentile 
(also known as the first quartile) of the 'AnnualIncome' values. 
The 25th percentile means that 25% of the data points in 
the 'AnnualIncome' column are below this value.'''

percent25_AI = round(loans['AnnualIncome'].quantile(0.25),2)

'''.quantile(0.75): This calculates the 75th percentile 
(also known as the third quartile) of the 'AnnualIncome' values. 
The 75th percentile means that 75% of the data points in the 'AnnualIncome' column are below this value.'''
percent75_AI = round(loans['AnnualIncome'].quantile(0.75),2)
```

```python
print(f' Mean = {mean_AI}')
print(f' Mode = {mode_AI}')
print(f' Median = {median_AI}')
print(f' 25th Percentile = {percent25_AI}')
print(f' 75th Percentile = {percent75_AI}')
```

```python
#Majority of the clients who get their loans approved are at what income level?

# Defining the function that will categorize income levels
def income_category(AnnualIncome):
    if AnnualIncome <= percent25_AI:
        return f'Lower Income (<= {percent25_AI})'
    elif AnnualIncome <= median_AI:
        return f'Lower-Middle Income (<= {median_AI})'
    elif AnnualIncome <= percent75_AI:
        return f'Upper-Middle Income (<= {percent75_AI})'
    else:
        return f'High Income (> {percent75_AI})'
    
loans['Income Category'] = loans['AnnualIncome'].apply(income_category)
loans[['Income Category', 'AnnualIncome']]
```

```python
# # Grouping by 'Income Category' and counting the number of people in each category
# applications_per_category = loans.groupby('Income Category', as_index= False)['Income Category'].size().sort_values('size', ascending= False)
# applications_per_category.head()
```

```python
# Grouping by 'Income Category' and counting the number of people in each category
applications_per_category = loans.groupby('Income Category')['Income Category'].size().reset_index(name= 'Number_of_Applicants')
applications_per_category = applications_per_category.sort_values('Number_of_Applicants', ascending= False)
applications_per_category.head()
```

```python
sns.barplot(data= applications_per_category, y= 'Income Category', x= 'Number_of_Applicants')
```

```python
loansApproval_per_category = loans.groupby(['Income Category', 'LoanApproved']).size().reset_index(name = 'Number_of_Applicants')
```

```python
loansApproval_per_category.head()
```

```python
plt.figure(figsize=(10,6))

myplot = sns.barplot(data= loansApproval_per_category, x= 'Income Category', y= 'Number_of_Applicants', hue= 'LoanApproved', errorbar= None)

for bar in myplot.containers:
    myplot.bar_label(bar, 
                     labels= [f'{v:0.0f}' for v in bar.datavalues],
                     padding= 2)

plt.title("Loan Approval rate spread Across Income Category",
          fontdict= {'fontsize' : 14,
                     'fontweight' : 'bold',
                     'color': '#89332a'})
plt.xlabel("Annual Income Category",
          fontdict= {'fontsize' : 12,
                     'fontweight' : 'bold',
                     'color': '#89332a'})
plt.xticks(ticks= [0,1,2,3], labels= ['High Income', 'Low Income', 'Low-Middle Income', 'High-Middle Income'])
plt.ylabel("Number of Applications",
          fontdict= {'fontsize' : 14,
                     'fontweight' : 'bold',
                     'color': '#89332a'})


myplot.legend(labels=['NA','A'], title = 'Approvals', loc = 'upper right')
plt.tight_layout()
plt.show()
```

```python
loansApproval_per_category_sorted = loansApproval_per_category.sort_values(['LoanApproved','Number_of_Applicants'], ascending= [False,True])
```

```python
loansApproval_per_category_sorted
```

```python
plt.figure(figsize=(10,6))

myplot = sns.barplot(data= loansApproval_per_category_sorted, x= 'Income Category', y= 'Number_of_Applicants',
                     hue= 'LoanApproved',
                     width= 0.6,
                     errorbar= None)

for bar in myplot.containers:
    myplot.bar_label(bar, 
                     labels= [f'{v:0.0f}' for v in bar.datavalues],
                     padding= 2)
for bars in myplot.containers:
    for bar in bars:
        bar.set_alpha(0.8)
plt.title("Loan Approval rate spread Across Income Category",
          fontdict= {'fontsize' : 14,
                     'fontweight' : 'bold',
                     'color': '#89332a'})
plt.xlabel("Annual Income Category",
          fontdict= {'fontsize' : 12,
                     'fontweight' : 'bold',
                     'color': '#89332a'})
plt.xticks(ticks= [0,1,2,3], labels= ['Low Income', 'Low-Middle Income', 'Upper-Middle Income', 'High Income'])
plt.ylabel("Number of Applications",
          fontdict= {'fontsize' : 14,
                     'fontweight' : 'bold',
                     'color': '#89332a'})


myplot.legend(labels=['NA','A'], title = 'Approvals', loc = 'upper right')
plt.tight_layout()
plt.show()
```
