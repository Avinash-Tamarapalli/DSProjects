    Correlation matrix and histograms

```python
from sklearn.preprocessing import LabelEncoder 

df['area error'] = LabelEncoder().fit_transform(df['area error'])

df.corr()
```

```python
df.iloc[:,0:4].corr(method= "spearman")
```

    Heatmap

```python
plt.figure(figsize= (14,9))

sns.heatmap(df.corr().abs(), cmap= 'coolwarm')

plt.show()
```

    Stacked Histogram

```python
df.head()
```

```python
sns.histplot(data= df, x= 'mean radius', hue= 'target', multiple= 'stack')
```

```python
sns.histplot(data= df, x= 'mean fractal dimension', hue= 'target', multiple= 'stack')	
```

```python
columns = df.select_dtypes(exclude= 'object').drop('target', axis= 1).columns
```

```python
columns
```

```python
fig, axs = plt.subplots(len(columns), 1, figsize=(5, 5 * len(columns)))


for i in range(len(columns)):
    column_name = columns[i]
    sns.histplot(data= df, x= column_name, hue= 'target', multiple= 'stack', ax= axs[i])

plt.show()
```
