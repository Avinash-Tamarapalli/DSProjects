```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

    Combining the Two DataFrames


```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
#To check Data Types
final_df.dtypes
```

    Exploring

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#Pie Chart - Top 3 Countries that uses Zomato

plt.pie(x=country_values[:3],labels= country_names[:3], autopct= '%.2f%%')
```

    Observation: 
The Maximum Transactions for Zomato comes from India, then followed by US and UK with 4.73% and 0.87% respectively

### 2. Explore about the Numerical Variables

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating' , 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

    Observations:
1. When Rating is between 4.5 to 4.9 -----------> Excellent
2. When Rating is between 4.0 to 4.4 -----------> Very Good
3. When Rating is between 3.5 to 3.9 -----------> Good
4. When Rating is between 2.5 to 3.4 -----------> Average
5. When Rating is between 1.8 to 2.4 -----------> Poor
6. When Rating is 0.0 --------------------------> No Rating

```python
# BAR PLOT


plt.figure(figsize=(12,6))

sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = 'Rating text' , palette='viridis', legend = False)
plt.show()
```

```python
# list(ratings['Rating color'].unique())
```

```python
plt.figure(figsize=(12,6))

sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = 'Rating color' , palette= list(ratings['Rating color'].unique()), legend = False)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(data = ratings, x = "Aggregate rating", y = "Rating Count", hue = 'Rating text' , palette= ['Black', 'Red', 'Orange', 'Yellow', 'Green', 'Blue'], legend = True)
plt.show()
```

    Observations:
1. There are high number of people who have NOT RATED
2. Second highest falls between 2.9 to 3.8

```python
# COUNT PLOT

plt.figure(figsize=(12,6))

sns.countplot(data = ratings, x = 'Rating color', palette= ['Black', 'Red', 'Orange', 'Yellow', 'Green', 'Blue'])

# the plot on y-axis shows frequency, don't confuse it with Rating count in ratings DF
plt.show()
```

```python
### Find the countries name that has given 0 rating
final_df[final_df['Aggregate rating'] == 0]['Country'].value_counts().reset_index()
```
